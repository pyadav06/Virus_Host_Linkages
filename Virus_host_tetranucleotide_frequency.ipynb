```python
import os 
from Bio import SeqIO
import pandas as pd

filepath='/content/Teo-Viral-contigs.fasta' 
window_size = 4       #Estimate the tetranucleotide frequency of viral contigs/genomes in a window size of 4.

a=['A','T','G','C']
lst=[]
s=""
for i in range(0,4):
  s=""
  s=s+a[i]
  for j in range(0,4):
      t=s+a[j]
      for k in range(0,4):
          u=t+a[k]
          for l in range(0,4):
              v=u+a[l]
              lst.append(v)
      
df = pd.DataFrame (lst, columns = ['Combination'])

for seq_record in SeqIO.parse(filepath, "fasta"):
#   print(seq_record.id)
    x=seq_record.seq
    dict1={}
    c=0
    for i in range(len(x) - window_size + 1):   #slide the window by 1 nucleotide to move on to next tetranucleotide combination.
        a=str(x[i: i + window_size])
        c=c+1
        if a in dict1:
            dict1[a]+=1
        else:
            dict1[a]=1

    #   print(len(seq_record))
    for k,v in dict1.items():
      dict1[k]=v/c
    df[seq_record.id]=df['Combination'].apply(lambda x:  dict1[x] if x in dict1 else 0)
virus=df
virus
```

```python
import os
from Bio import SeqIO
import pandas as pd

def get_df(dir):              #Calculate the tetranucleotide frequency for the metagenome assembled genomes from metagenome which served as viral hosts.
      a=['A','T','G','C']
      lst=[]
      s=""
      for i in range(0,4):
        s=""
        s=s+a[i]
        for j in range(0,4):
            t=s+a[j]
            for k in range(0,4):
                u=t+a[k]
                for l in range(0,4):
                    v=u+a[l]
                    lst.append(v)
      
      df = pd.DataFrame (lst, columns = ['Combination'])

      directories = os.listdir(dir)
      
      # This would print all the files and directories
      for file in directories:
          filename=str(file).replace(".fa","")
          window_size = 4
          dict1={}
          c=0
          filepath=dir+"/"+str(filename)+".fa"
          boolean=os.path.exists(filepath)
          if boolean==False:
            continue
          for seq_record in SeqIO.parse(filepath, "fasta"):
          #   print(seq_record.id)
              x=seq_record.seq
              for i in range(len(x) - window_size + 1):
                  a=str(x[i: i + window_size])
                  c=c+1
                  if a in dict1:
                      dict1[a]+=1
                  else:
                      dict1[a]=1

          #   print(len(seq_record))
          for k,v in dict1.items():
            dict1[k]=v/c
          df[filename]=df['Combination'].apply(lambda x:  dict1[x] if x in dict1 else 0)
      return df 
```

```python
Bacteria=get_df("/content/MAGs/")
Bacteria
```

```python
from scipy.spatial import distance
from scipy.stats import pearsonr
virus_list=list(virus.columns[1:]).  #calculate the Euclidean-distance, mean asbolute error (MAE), and pearson's coefficient to link viruses with their putative hosts.
bacteria_list=list(Bacteria.columns[1:])
lst_corr=[]

for i in virus_list:
  for j in bacteria_list:
    bacteria_series=Bacteria[j]
    virus_series=virus[i]
    dst = distance.euclidean(bacteria_series, virus_series)
    Absolute = mae(bacteria_series, virus_series)
    corr=pearsonr(bacteria_series, virus_series)[0]
    lst_corr.append((j,i,dst,Absolute,corr))
result=pd.DataFrame(lst_corr,columns = ['MAG','Virus','Euclidean-distance', 'MAE', 'pearson'])
result   
```
